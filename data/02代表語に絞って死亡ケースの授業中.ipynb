### 死亡ケースのさらに授業・休憩時間に絞って

```python
import pandas as pd

# データ読み込み
df_shibou = pd.read_excel("変更後_sibou(2025.01.31).xlsx")

# 確認：対象となる主な列
print(df_shibou.columns.tolist())
df_shibou.head()
```

```python
# 「分類カテゴリ_場合別2」が「授業と休憩時間」のものだけに絞る
df_filtered = df_shibou[df_shibou["分類カテゴリ_場合別2"] == "授業と休憩時間"].copy()

# フィルタできているか確認（行数）
print(len(df_filtered))
```

```python
# 欠損を空文字で埋めた上で、4列の情報を結合
docs = (
    "状況: " + df_filtered["災害発生時の状況"].fillna("") + "。 " +
    "種別: " + df_filtered["種別"].fillna("") + "。 " +
    "場合: " + df_filtered["場合別2"].fillna("") + "。 " +
    "競技: " + df_filtered["競技種目"].fillna("")
).astype(str).tolist()

# 結果の確認（例：先頭3件）
for d in docs[:3]:
    print(d)
```

```python
from fugashi import Tagger  # ← GenericTaggerではなくこちら

tagger = Tagger()  # ← これでOK（unidic-lite を自動で使う）

def tokenize(text):
    return [
        word.surface 
        for word in tagger(text) 
        if word.feature[0] not in ["記号", "助詞", "助動詞", "接続詞", "連体詞"]
    ]

tokenized_docs = [" ".join(tokenize(doc)) for doc in docs]

# 結果の確認（例: 先頭3件）
for t in tokenized_docs[:3]:
    print(t)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np

# ① TF-IDF ベクトル化（必要に応じてstop_wordsを指定してもOK）
# 不要語のリスト（必要に応じて増やしてください）
custom_stopwords = ["する", "いる", "ため", "こと", "ある", "なる", "及び", "および", "において", 
                    "により", "など", "もの", "よう", "そして", "また", "場合", "時", "時間", 
                    "以外", "確認", "管理", "実施", "状況", "事故"]

# TF-IDFベクトル化
vectorizer = TfidfVectorizer(stop_words=custom_stopwords)
X = vectorizer.fit_transform(tokenized_docs)

# ② クラスタ数を指定してKMeansでクラスタリング
k = 10  # 例：5クラスタ
kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
kmeans.fit(X)

# ③ 各クラスタの代表語（TF-IDFの高い語）を表示
terms = vectorizer.get_feature_names_out()
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

for i in range(k):
    print(f"\nクラスタ {i} の特徴語:")
    for ind in order_centroids[i, :20]:
        print(" ", terms[ind])
```

### クラスタ分析の可視化について（代表語に絞って可視化する方法）

```python
cluster_top_terms = {}  # {cluster_id: [term1, term2, ...]}
for i in range(k):
    top_terms = [terms[ind] for ind in order_centroids[i, :10]]
    cluster_top_terms[i] = top_terms
```

```python
# 全文トークンから「共起関係（リンク）」を作成
from collections import defaultdict
import itertools

# 各文書に出現する単語リスト
cooccurrence = defaultdict(int)

for doc in tokenized_docs:
    tokens = set(doc.split())
    for w1, w2 in itertools.combinations(tokens, 2):
        if w1 != w2:
            pair = tuple(sorted([w1, w2]))
            cooccurrence[pair] += 1
```

```python
# クラスタごとに代表語＋共起語のノード・リンク構築：クラスタ1用
cluster_id = 1
top_words = set(cluster_top_terms[cluster_id])
min_count = 3  # 閾値（必要に応じて調整）

nodes = []
edges = []
added_words = set()

# 代表語ノード追加
for word in top_words:
    nodes.append({"id": word, "value": 5, "cluster": cluster_id})
    added_words.add(word)

# 共起語とリンク追加
for (w1, w2), freq in cooccurrence.items():
    if freq >= min_count:
        if w1 in top_words or w2 in top_words:
            for w in [w1, w2]:
                if w not in added_words:
                    nodes.append({"id": w, "value": 2, "cluster": cluster_id})
                    added_words.add(w)
            edges.append({"source": w1, "target": w2, "value": freq})
```

```python
import json
network = {"nodes": nodes, "links": edges}

with open("cluster1_network_expanded.json", "w", encoding="utf-8") as f:
    json.dump(network, f, ensure_ascii=False, indent=2)
```

```python
import json

# 1. 元のクラスタ1データ読み込み
with open("cluster1_network_expanded.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 2. ストップワード（不要語）リストを定義
stopwords = {
    ":","：","さ","ところ", "大", "２", "し", "下", "（", "い", "、", "」", "１", "よう", "あっ", "あり",
    "こと", "ため", "もの", "さん", "等", "ほか", "「", "）", "。", "・", "…", "っ", "々",
    "の", "が", "を", "に", "と", "で", "は", "も", "や", "より", "へ", "～",
    "する", "いる", "ある", "なる", "そして", "また", "場合", "時", "時間",
    "確認", "管理", "実施", "状況", "事故", "及び", "および"
}
stopwords = {w.strip() for w in stopwords}  # 念のため空白除去

# 3. 不要語を含むノードとリンクを除去
data["nodes"] = [node for node in data["nodes"] if node["id"].strip() not in stopwords]
valid_ids = {node["id"] for node in data["nodes"]}
data["links"] = [
    link for link in data["links"]
    if link["source"] in valid_ids and link["target"] in valid_ids
]

# 4. 代表語リスト（中心に置きたい語）
hub_terms = {"給食", "児童", "窒息"}

# 5. 共起頻度のしきい値を設定
THRESHOLD = 10

# 6. リンクを代表語に限定し、しきい値以上のみに絞る
filtered_links = [
    link for link in data["links"]
    if (link["source"] in hub_terms or link["target"] in hub_terms)
    and link["value"] >= THRESHOLD
]

# 7. 使用されているノードだけを残す
target_ids = {link["source"] for link in filtered_links} | {link["target"] for link in filtered_links}
filtered_nodes = [node for node in data["nodes"] if node["id"] in target_ids]

# 8. JSON形式でまとめて保存
filtered_data = {
    "nodes": filtered_nodes,
    "links": filtered_links
}

with open("cluster1_filtered.json", "w", encoding="utf-8") as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=2)

print("✅ cluster1_filtered.json を保存しました")
```
